```python
# Extrnal poison exon data
## Leclair et al 2020
leclair_data = "../data/External_Data/hg38/SRSF_PoisonExons_Leclair_etal_2020_MolCell.txt"
## Felker et al 2023
felker_data = "../data/External_Data/hg38/Felker2023SuppelementaryTable1_hg38_PE_cassettes_and_elements_annotated.tsv"
## Shane Lawson shared data 2026
lawson_data = "../data/External_Data/hg38/NMD_KD-dPSI_hg38.bed"
lawson_NMDexons_data = "../data/External_Data/hg38/nmd_exons (1).bed"
lawson_curNMDcons_data = "../data/External_Data/hg38/curated_nmd_cons (1).tsv"
lawson_valNMDcons_data = "../data/External_Data/hg38/validated_nmd_cons (1).tsv"
```

```python

# Read in data
leclair = pd.read_csv(leclair_data, index_col=False, sep = '\t')
felker = pd.read_csv(felker_data,index_col=False, sep = '\t')
lawson = pd.read_csv(lawson_data, index_col=False, sep = '\t', header = None)
lawson_NMDexons = pd.read_csv(lawson_NMDexons_data, index_col=False, sep = '\t', header = None, names=["gene_symbol","PE_chromosome","PE_start","PE_end","PE_strand","Lawson_NMD_Exons_Score","PE_type"])
lawson_curNMDcons = pd.read_csv(lawson_curNMDcons_data, index_col=False, sep = '\t', header = None, names=["gene_symbol","PE_chromosome","PE_start","PE_end","PE_strand","Lawson_CurNMD_Exons_Score","PE_type"])
lawson_valNMDcons = pd.read_csv(lawson_valNMDcons_data, index_col=False, sep = '\t')
```

```python
# Clean Leclair
leclair['PE_identifier'] = leclair['PoisonExon_hg38'].str.replace(",","")
leclair[['PE_chromosome','PE_start','PE_end','PE_strand']] = leclair['PE_identifier'].str.split(":|-",regex=True, expand = True, n = 3)
leclair['source'] = "Leclair et al. 2020"
leclair = leclair.rename(columns = {"Gene": "gene_symbol"})
```

```python
# Clean Lawson
lawson_NMDexons['source'] = "Shane Lawson from annotation"
lawson_NMDexons.loc[lawson_NMDexons['PE_end'].isna(), 'PE_end'] = 39685690
lawson_curNMDcons['source'] = "Shane Lawson conserved curated"
lawson_valNMDcons = lawson_valNMDcons.rename(columns = {"Gene": "gene_symbol", "Exon": "PE_exon", "hg38_coord": "PE_identifier", "Type": "PE_Type", "Source":"source_primary"})
lawson_valNMDcons[['PE_chromosome','PE_start','PE_end']] = lawson_valNMDcons['PE_identifier'].str.split(":|-",regex=True, expand = True, n = 3)
```
